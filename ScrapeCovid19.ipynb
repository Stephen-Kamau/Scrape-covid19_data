```python
#will be scapping data from worldmeter dashboeard
import pandas as pd
import urllib
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.worldometers.info/coronavirus/'
web_url = requests.get(url).text
soup = bs(web_url , 'html.parser')
```
